```python
# 1. Write a program to check if a number is Positive, Negative, or Zero.
number=int(input('enter your number:'))
if number >0:
    print('positive')
elif number<0:
    print('negative')
else:
    print('zero')
```

```python
# 2. Write a program to check if a number is Even or Odd.
number=int(input('enter your number'))
if number % 2==0:
    print('even number')
else:
    print('odd number')
```

```python
# 3. Write a program to find the greater number between two numbers.
number_1=int(input('enter your first number'))
number_2=int(input('enter your second number'))
if number_1>=number_2:
    print(number_1,'is great number',number_2,'is small number')
else:
    print(number_2,'is great number',number_1,'is small number')
```

```python
# 4. Write a program to take marks from the user and print the Grade using:
# A (90), B (75), C (60), D (40), Fail (below 40).

marks=int(input('enter your marks'))
if marks>=90:
    print('Grade A')
elif marks>=75:
    print('Grade B')
elif marks>=60:
    print('Grade C')
elif marks>=40:
    print('Grade D')
else:
    print('Fail')
```

```python
# 5. Write a program to take age from the user and print whether the user is a Child, Teenager, Adult, or Senior Citizen.

age=int(input('enter your age'))

if age>=70:
    print('Senior Citizen')
elif age>=18:
    print('Adult')
elif age>=13:
    print('Teenager')
else:
    print('Child')
```

```python
# 6. Write a program to input temperature and print weather suggestion:
# Very Hot, Hot, Warm, Pleasant, Cold, Very Cold.

temp=float(input('temp'))
if temp>=52:
    print('today is very hot')
elif temp>=40:
    print('today is hot')
elif temp>=30:
    print('''it's okay today''')
elif temp>=20:
    print('today is pleasent day')
elif temp>=14:
    print('today is cold')
else:
    print('today is very cold')
```

```python
###### 8. Write a program to check Password Strength as Weak, Moderate, or Strong using length of password.
pswd = input('Enter your password:')

if len(pswd) < 4:
    print('Weak Password')
elif len(pswd) >= 4 and len(pswd) < 7:
    print('Moderate Password')
else:
    print('Strong Password')
```

```python
# 9. Write a program to check Mobile Recharge Plan based on amount entered (like ₹99, ₹199, ₹399)
recharge = int(input('Enter your recharge amount: '))

if recharge == 99:
    print('14GB total data (0.5GB/day), unlimited calls, 300 SMS, validity: 28 days')

elif recharge == 199:
    print('1.5GB/day data, unlimited calling, 100 SMS/day, JioTV + JioCinema + JioCloud, validity: 28 days')

elif recharge == 399:
    print('2.5GB/day data, unlimited calling, 100 SMS/day, Jio Apps access, validity: 28 days')

else:
    print('Invalid Recharge Amount. Please enter 99, 199, or 399.')
```
