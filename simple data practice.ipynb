```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\shafqat Ali\OneDrive\Documents\dummydata.csv")
df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
col=['Product','Category']
for i in col:
    df[i]=df[i].str.strip()
    df[i] = df[i].str.strip().str.lower()
```

```python
 # Kaun-sa product sabse zyadh sales laata hai?
df.loc[df['Total_Sales'].id]
```

```python
 # Kaun-sa product sabse kam sales laata hai?
df.loc[df['Total_Sales'].idxmin(),'Product']
```

```python
# Kis category me total quantity sabse zyada bikti hai?
df.loc[df['Quantity'].idxmax(),'Category']
```

```python
# Kis product ka price highest hai?
df.loc[df['Price'].idxmax(),'Product']
```

```python
# Kis product ka rating lowest hai?
df.loc[df['Rating'].idxmin(),'Product']
```

```python
# Category-wise total sales kitni hai?
df.groupby('Category')['Total_Sales'].sum()
```

```python
df.groupby('Product')['Total_Sales'].sum()
```

```python
# Category-wise average rating kya hai?
df.groupby('Category')['Rating'].mean()
```

```python
# Price aur Rating ka relation kya dikh raha hai?
df[['Price','Rating']].corr()
```

```python
# Kis category ka average price sabse zyada hai?
result = df.groupby('Category')['Price'].mean().sort_values(ascending=False)

print(result)
```
